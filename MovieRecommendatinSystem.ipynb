```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies=movies.drop(columns=['Director','Star1','Star2','Star3','Star4','Poster_Link','Released_Year','Certificate','Runtime','IMDB_Rating','Meta_score','No_of_Votes','Gross'])
```

```python
movies['tags']=movies['Genre']+movies['Overview']
```

```python
movies.head()
```

```python
movies=movies.drop(columns=['Genre','Overview'])
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Initialize CountVectorizer with a limit on max features and removing English stop words
cv =CountVectorizer(max_features=1000,stop_words='english')
# Fit and transform the movie descriptions (assuming you have a DataFrame `movies` with a 'Description' column)
vec = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
print(vec.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Calculate cosine similarity matrix
sim =cosine_similarity(vec)
```

```python
def recommend(movie):
  index= movies[movies['Series_Title']==movie].index[0]
  dist=sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
  for i in dist[1:9]:
    print(movies.iloc[i[0]].Series_Title)

```

```python
recommend('The Godfather')
```
